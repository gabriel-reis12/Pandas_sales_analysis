1.   Criar um novo data frame
2.   Descobrir a média de vendas e porcentagem por categoria

1 =

```python
df = dados[['City', 'Category', 'Sub-Category', 'Sales']]
colunas = {'City': 'Cidade', 'Category': 'Categoria', 'Sub-Category':'Sub_Categoria', 'Sales':'Vendas'}
df.rename(columns = colunas, inplace = True)
df
```

2 =

```python
media_de_vendas = df.groupby('Categoria')['Vendas'].mean().round(2)
```

```python
novo_df = pd.DataFrame(media_de_vendas)
novo_df.rename(columns= {'Vendas': 'Media de vendas'}, inplace = True)
```

```python
novo_df
```

```python
total_de_vendas = novo_df['Media de vendas'].sum()
total_de_vendas
```

```python
novo_df['% de Vendas'] = novo_df['Media de vendas'].apply(lambda x: (x / total_de_vendas) * 100).round(2)
```
